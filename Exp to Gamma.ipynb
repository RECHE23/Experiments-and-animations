```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
x = np.linspace(-5, 5, 100)
P = stats.expon.pdf(x, 0.3, 0.25)
Q = stats.expon.pdf(x, 0.1, 0.5)
R = stats.expon.pdf(x, 0.5, 0.75)
S = stats.expon.pdf(x, 0.1, 0.3)
T = stats.expon.pdf(x, 0.2, 0.6)
U = stats.expon.pdf(x, 0.7, 0.8)
PQR = P + Q + R + S + T + U

#plt.plot(x, P)
#plt.plot(x, Q)
#plt.plot(x, R)
plt.plot(x, PQR)
plt.title('Exp')
plt.legend(('P', 'Q', 'PQ'), loc='upper left', shadow=True);
```

```python
x = np.linspace(-5, 5, 100)
V = np.zeros(x.shape)
for i in range(1000):
    V += stats.expon.pdf(x, *np.random.rand(2))
    
plt.plot(x, V)
plt.title('Exp')
plt.legend(('V'), loc='upper left', shadow=True);
```

```python
x = np.linspace(-5, 5, 100)
V = np.zeros(x.shape)
for i in range(1000000):
    V += stats.expon.pdf(x, *np.random.rand(2))
    
plt.plot(x, V)
plt.title('Exp')
plt.legend(('V'), loc='upper left', shadow=True);
```

```python
V_cum = np.cumsum(V)
V_cum = V_cum/V_cum[-1]

plt.plot(x, V_cum)
plt.title('Exp')
plt.legend(('V_cum'), loc='upper left', shadow=True);
```

```python
V_ = np.ediff1d(V_cum, to_begin=V_cum[0])

plt.plot(x, V_)
plt.title('Exp')
plt.legend(("V'"), loc='upper left', shadow=True);
```

```python
mean = np.sum(V_ * x)
var = np.sum(V_ * (V_ - mean) ** 2)

mean, var
```
